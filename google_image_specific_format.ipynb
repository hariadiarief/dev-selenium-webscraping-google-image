```python
# Set up the Selenium driver
options = uc.ChromeOptions()
driver = uc.Chrome(options=options)

# save results
results = []

# Scrap
for query in search_queries:
    driver.get("https://www.google.com/imghp")
 
    search_box = driver.find_element(By.XPATH, '//textarea[@name="q"]')
    search_box.send_keys(f"site:www.tokopedia.com {query}")
    search_box.send_keys(Keys.RETURN)

    wait = WebDriverWait(driver, 10)

    allowed_format = [".jpg" , '.png'] # list of expected data format
    
    # result form
    result = {
        "name": query,
        "image": "",
    }


    # get list of images 
    try:
        list_images = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, ".rg_i.Q4LuWd")))
    except Exception as e:
        print(e)

    # get higer res of each list images from its detail
    for index, image in enumerate(list_images):
        # if index == 10:
        #     break 
        # image.click() 

        try:
            image.click()
        except Exception as e:
            break
        
        try:
            img_url = wait.until(EC.visibility_of_element_located((By.CSS_SELECTOR, ".sFlh5c.pT0Scc.iPVvYb"))).get_attribute("src")

            if any(substring in img_url for substring in allowed_format):
                result["image"] = img_url
                break   
        except Exception as e:
            result["image"] = ""
            print(e)

    results.append(result) 

# Close the browser
driver.quit()

print("---DONE--")
```
